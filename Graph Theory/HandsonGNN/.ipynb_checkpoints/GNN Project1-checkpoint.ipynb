### showing sample molecules

```python
data
```

```python
sample_smiles = data['smiles'][4:30].values
```

```python
sample_mols=[Chem.MolFromSmiles(smiles) for smiles in sample_smiles]
```

```python
Grid=Draw.MolsToGridImage(sample_mols, molsPerRow=4, subImgSize=(200,200))
```
